# DRaCOoN tutorial

## Importing the necessary packages

```python
from app.dracoon import dracoon
from validation.simulated_data.GRN_simulator import GRN_simulator
```

## Simulating data

```python
simulation = GRN_simulator(n_genes=50, n_tfs=20, n_samples=300,
                           n_shutdowns=1, n_shifts =1,
                           n_inversions=1, n_uncorrelateds=1,
                           proportion_group_case=0.5,
                           simdata_noise=0.5, perturbation_noise=0.5, mean_expression=2, shift_alpha=2,
                           ingroup_perturbed_proportion=1)
simulation.generate_DRN()
simulation.plot_clustermap()
```

### Checking the simulated data

Simulated expression data, with the genes in columns and the samples in the rows.

```python
print(simulation.simulated_data.shape)
simulation.simulated_data.head()
```

Condition dataframe, with the samples in the row indices and the conditions in the columns 'condition'. This column is required to run DRaCOoN.

```python
print(simulation.condition_data.condition.value_counts())
simulation.condition_data.head()
```

Structure dataframe, the simulated relationships between tfs and tfs and tfs and genes in the columns 'source' and 'target', respectively. The interactive network plot shows the structure of the simulated network, with the red genes being the perturbed ones.

```python
print(simulation.structure.head())
simulation.plot_pyvis().show('structure.html')
```

Reference network, or ground truth for validation. The reference differential network is composed of the edges in the structure network that contain at least one of the perturbed genes.

```python
print(simulation.refnet.head())
simulation.plot_pyvis(relationships=simulation.refnet).show('refnet.html')
```
